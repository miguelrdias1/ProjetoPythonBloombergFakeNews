```python
import pandas as pd
import sklearn #importando biblioteca de skillLearn
from sklearn.model_selection import train_test_split #importando train_test_split de skill learning
from sklearn.feature_extraction.text import TfidfVectorizer ##Importando TfidVectorizer de skill learning
from sklearn.linear_model import PassiveAggressiveClassifier ##Importando o modelo de IA que vai deicidir se é falsa ou real
from sklearn.metrics import accuracy_score, confusion_matrix #importando os testes pra IA realizar
import joblib
```

```python
df = pd.read_csv("/content/Data/pre-processed.csv")
##Importando os arquivos de tabelas pela maquina virtual do Google
```

```python
#Checando o data frame

print(df.shape)        # quantas linhas/colunas
print(df.head())        #Quais os dominios
print(df["label"].value_counts())  # Vendo se bate o numero de noticias falsas com reais
print(df.isnull().sum())           # checar nulos
```

```python
##Criando uma nova coluna, que tem 0 nas tuplas de noticias falsas e 1 nas de verdadeiras.
df["label_num"]=df["label"].map({"fake":0, "true":1})

X_train, X_test, y_train, y_test = train_test_split( ##no teste, o modelo recebe y com 1 ou 0 (verdadeira ou falsa) e o texto como X,e compara se o valor que ele
                                                     ##ligou a X (verdadeiro ou) bate com a verdadeira(Y)
    df["preprocessed_news"], df["label_num"], test_size=0.2, random_state=42 ##separando 20% das labels para serem usadas como testes e 80% como treino
)


tfidf = TfidfVectorizer(max_df=0.7)
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf  = tfidf.transform(X_test)
##Vai converter as palavras em parametros, e ignorar os que aparecem em 80% de todas as noticias

pac = PassiveAggressiveClassifier(max_iter=50)
pac.fit(X_train_tfidf, y_train)

y_pred = pac.predict(X_test_tfidf)
print(f"Acurácia: {accuracy_score(y_test, y_pred):.4f}")
print (confusion_matrix(y_test, y_pred))
```

```python
joblib.dump(pac, "modelo.pkl")
joblib.dump(tfidf, "tfidf.pkl")
```
